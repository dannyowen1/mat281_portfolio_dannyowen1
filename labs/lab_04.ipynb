En la carpeta data se encuentra el archivo `nba.db`, el cual muestra informacion básica de algunos jugadores de la NBA.

```python
from sqlalchemy import create_engine
import pandas as pd
import os  

# Crear conector
conn = create_engine(os.path.join('sqlite:///','data', 'nba.db'))
```

```python
# funcion de consultas
def consulta(query,conn):
    return pd.read_sql_query(query, con=conn)
```

```python
# ejemplo
query = """
SELECT * FROM player
"""
consulta(query,conn)
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos mediante código **SQL**. Para cumplir con este objetivo debe resolver las siguientes problemáticas:

1. Mostrar las primeras 5 filas

```python
query = """
SELECT * FROM player LIMIT 5;
"""
consulta(query,conn)
```

2. Seleccionar los valores únicos de la columna `position`.

```python
query = """
SELECT DISTINCT position FROM player 
"""
consulta(query,conn)
```

3. Seleccionar y cambiar el nombre de la columna `name` por `nombre` 

```python
query = """
SELECT name as nombre FROM player 
"""
consulta(query,conn)
```

4. Determinar el tiempo (en años) de cada jugador en su posición

```python
# incluiremos en nombre del jugador para una mejor visualizacion
query = """
SELECT name,year_end-year_start FROM player 
"""
consulta(query,conn)
```

5. Encontrar el valor máximo de la columna `weight` por cada  valor de la columna `position`

```python
# incluiremos el dato 'position' para una mejor visualizacion del valor entregado
query = """
SELECT position,MAX(weight) FROM player GROUP BY position 
"""
consulta(query,conn)
```

6. Encontrar el total de jugadores por cada valor de la columna `year_start`

```python
# incluiremos el dato del año para una mejor visualizacion
query = """
SELECT year_start,COUNT() FROM player GROUP BY year_start 
"""
consulta(query,conn)
```

7. Encontrar el valor mínimo, máximo y promedio de la columna `weight` por cada valor de la columnas `college`

```python
# incluiremos el dato del 'college' para saber de cual se trata
query = """
SELECT college,MAX(weight),MIN(weight),AVG(weight) FROM player GROUP BY college 
"""
consulta(query,conn)
```

8. Filtrar por aquellos jugadores que cumplan con :
* Para la columna `year_start` tienen un valor mayor 1990 y menor a 2000
* Para la columna `position` tienen un valor de `C`,`C-F` o `F-C`
* Para la columna `college` tienen un valor distinto de `Duke University`


```python
# incluiremos el dato del nombre para saber de que jugador se trata, para las filtraciones antes mencionadas
query = """
SELECT name,year_start,position,college FROM player 
WHERE year_start>1990 AND year_start <2000 AND position='C' AND NOT college='Duke University'
OR year_start>1990 AND year_start <2000 AND position='C-F' AND NOT college='Duke University'
OR year_start>1990 AND year_start <2000 AND position='F-C' AND NOT college='Duke University'
"""
consulta(query,conn)
```

```python
# manera alternativa mas corta
# incluiremos el dato del nombre para saber de que jugador se trata, para las filtraciones antes mencionadas
query = """
SELECT name,year_start,position,college FROM player 
WHERE year_start>1990 AND year_start <2000 AND NOT college='Duke University' AND (position='C' OR position='C-F' OR position='F-C')

"""
consulta(query,conn)
```

9. Crear dos conjuntos de datos y juntarlos en una misma *query*. Las condiciones de cada uno de los cojunto de datos son:

* **df1**:
    * Para la columna `year_start` tienen un valor mayor 1990 y menor a 2000
    * Para la columna `position` tienen un valor de `C`,`C-F` o `F-C`
   
* **df2**:
    * Para la columna `year_end` tienen un valor menor a 2000
    * Para la columna `position` tienen un valor de `G`o `F` 

```python
# Creacion lista df1 (omitir el OperationalError entregado abajo, puesto que se volvio a callear la dataframe ya hecha)
query = """
CREATE TABLE df1 AS
    SELECT *
    FROM player
    WHERE year_start > 1990 AND year_start < 2000
    AND (position = 'C' OR position = 'C-F' OR position = 'F-C');
"""
consulta(query,conn)
```

```python
# Creacion lista df2 (omitir el OperationalError entregado abajo, puesto que se volvio a callear la dataframe ya hecha)
query = """
CREATE TABLE df2 AS
    SELECT *
    FROM player
    WHERE year_end < 2000
    AND (position = 'F' OR position = 'G');
"""
consulta(query,conn)
```

```python
query = """
SELECT * FROM df1 UNION ALL SELECT * FROM df2
"""
consulta(query,conn)
```
