Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
print(df['total_profit'])
plt.figure(figsize=(10, 6))
#palette = sns.color_palette("Set2", 4)
sns.lineplot(
    x='month_number',
    y='total_profit',
    data=df,
    ci = None,
    #palette=palette
    color = 'Yellow'
)
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='month_number',
    y='total_profit',
    data=df,
    #palette=palette
    color = 'Red'
)
plt.show()
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
print(df[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']])
sns.lineplot(data=df.drop(['month_number','total_units','total_profit'],axis=1))
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
print(df[['facecream','facewash']])

plt.figure(figsize=(10, 6))
sns.barplot(
    x='month_number',
    y='facecream',
    data=df
)
print(plt.show())

plt.figure(figsize=(10, 6))
sns.barplot(
    x='month_number',
    y='facewash',
    data=df
)
print(plt.show())
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
# Pre-format DataFrame
stats_df = df.drop(['month_number', 'total_units', 'total_profit'], axis=1)
 
# New boxplot using stats_df
plt.figure(figsize=(10, 6))
sns.boxplot(data=stats_df)
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
print(df.sum())
```

```python
totals=df.sum()
t=[totals[1],totals[2],totals[3],totals[4],totals[5],totals[6]]
colors = sns.color_palette('bright')[0:6]
labels = ['facewash','facecream','toothpaste','bathingsoap','shampoo','moisturizer']
plt.pie(t, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
